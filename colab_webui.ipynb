```
import threading
import os
import time
import shutil
source_directory = "/content/GPT-SoVITS/output,/content/GPT-SoVITS/GPT_weights,/content/GPT-SoVITS/SoVITS_weights"#@param {type:"string"}
destination_directory = "/content/drive/MyDrive/gpt-SoVITS/b\u7AD9\u67D0up"#@param {type:"string"}
def backup_files(source, destination):
    """
    Function to backup files from source to destination.
    """
    for filepath in source:
        name = os.path.basename(filepath)
        des_path = os.path.join(destination,name)
        if os.path.exists(des_path):
            shutil.rmtree(des_path)
        if not os.path.exists(filepath):
            print(f"不存在:{filepath}")
            continue
        shutil.copytree(filepath, des_path)
        print(f"{name}:Backup completed!")

# 创建并启动备份文件的线程
source_file = []
if source_directory.split(",") == None:
    source_file.append(source_directory)
else:
    source_file = source_directory.split(",")

backup_files(source_file, destination_directory)
print("Backup is running in the background...")

```
